## Windows

```python
re.findall(r'Windows\s\d+|Mac OS|DOS',x[0].text)
```

```python
import numpy as np
```

```python
windows=[]
for i in x:
    a= re.findall(r'Windows\s\d+|Mac OS|DOS',i.text)
    if len(a)>0:
        windows.append(a[0])
    else:
        windows.append(np.nan)
windows
```
